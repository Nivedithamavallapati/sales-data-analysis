```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
```

```python
df.dtypes
```

```python
Q1 = df['Sales'].quantile(0.25)

Q3 = df['Sales'].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR

upper_limit = Q3 + 1.5 * IQR

print("Lower Limit :", lower_limit)

print("Upper Limit :", upper_limit)
```

```python
df_no_outliers = df[
    (df['Sales'] >= lower_limit) &
    (df['Sales'] <= upper_limit)
]
```

```python
print("Original Shape :", df.shape)

print("New Shape :", df_no_outliers.shape)
```

## DATA VISUALIZATION & ANALYSIS

# BAR CHART (REGION)

```python
plt.figure(figsize=(8,5))

sns.barplot(x='Region', y='Sales', data=df, errorbar=None)

plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")

plt.show()

print("Key Finding:")
print("West and East regions generated higher sales compared to other regions.")
```

# BAR CHART (CATEGORY)

```python
plt.figure(figsize=(8,5))

sns.barplot(x='Category', y='Sales', data=df, errorbar=None)

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.show()

print("Key Finding:")
print("Technology category contributed the highest sales among all categories.")
```

# LINE CHART

```python
monthly_sales = df.groupby(df['Order Date'].dt.month)['Sales'].sum()

plt.figure(figsize=(8,5))

monthly_sales.plot(marker='o')

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()

print("Key Finding:")
print("Sales fluctuate throughout the year with some months showing peak performance.")
```

# PIE CHART

```python
colors = sns.color_palette("Blues", 3)

df['Category'].value_counts().plot.pie(
    autopct='%1.1f%%',
    colors=colors,
    startangle=90
)

plt.title("Category Distribution")

plt.ylabel("")

plt.show()

print("Key Finding:")
print("Office Supplies and Technology occupy major portions of total sales categories.")
```

# TOP STATES

```python
top_states = df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))

top_states.plot(kind='bar')

plt.title("Top 10 States by Sales")
plt.xlabel("State")
plt.ylabel("Sales")

plt.show()

print("Key Finding:")
print("Certain states contribute significantly higher sales compared to other states.")
```

# HEATMAP

```python
plt.figure(figsize=(8,5))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap="Blues",
    linewidths=0.5
)

plt.title("Correlation Heatmap")

plt.show()

print("Key Finding:")
print("Correlation heatmap shows relationships between numerical features.")
```

# SCATTER PLOT

```python
avg_sales = df.groupby('State')['Sales'].mean().reset_index()

plt.figure(figsize=(12,6))

plt.scatter(avg_sales['State'], avg_sales['Sales'])

plt.xticks(rotation=90)

plt.xlabel("State")
plt.ylabel("Average Sales")

plt.title("Average Sales by State")

plt.show()

print("Key Finding:")
print("Average sales vary significantly across different states.")
```

# HISTOGRAM

```python
plt.figure(figsize=(10,5))

sns.histplot(
    df['Sales'],
    bins=30,
    kde=True,
    color='steelblue'
)

plt.title(
    "Distribution of Sales",
    fontsize=16,
    fontweight='bold'
)

plt.xlabel("Sales")

plt.ylabel("Frequency")

plt.grid(alpha=0.3)

plt.show()

print("Key Finding:")
print("Most sales values are concentrated in lower ranges with few high-value transactions.")
```

# BOX PLOT

```python
plt.figure(figsize=(10,5))

sns.boxplot(
    x=df['Sales'],
    color='lightblue',
    linewidth=2
)

plt.title(
    "Sales Boxplot Analysis",
    fontsize=16,
    fontweight='bold'
)

plt.xlabel("Sales")

plt.grid(alpha=0.3)

plt.show()

print("Key Finding:")
print("Boxplot reveals presence of outliers in sales data.")
```
